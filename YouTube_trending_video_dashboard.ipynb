```python

import zipfile
import os

# Extract the uploaded zip
zip_path = next(iter(uploaded))  # Gets the uploaded file name
with zipfile.ZipFile(zip_path, 'r') as zip_ref:
    zip_ref.extractall("/content/youtube_data")

# List files to locate the CSV
for root, dirs, files in os.walk("/content/youtube_data"):
    for file in files:
        if file.endswith(".csv"):
            csv_path = os.path.join(root, file)

print(f"CSV file found at: {csv_path}")
```

```python

import pandas as pd

# Load CSV into DataFrame
df = pd.read_csv(csv_path)
df.dropna(how='all', inplace=True)

# Convert trending_date if available
if 'trending_date' in df.columns:
    df['trending_date'] = pd.to_datetime(df['trending_date'], errors='coerce')

df.head()
```

```python

import plotly.express as px

# Top 10 Most Viewed Videos
top_views = df.sort_values(by='views', ascending=False).head(10)

# Define a custom color palette
custom_colors = ['#FF6B6B', '#6BCB77', '#4D96FF', '#FFD93D', '#FF9CEE',
                 '#845EC2', '#00C9A7', '#FFC75F', '#F9F871', '#D65DB1']

# Plot with custom colors
fig1 = px.bar(
    top_views,
    x='title',
    y='views',
    color='channelName',
    title='🔝 Top 10 Most Viewed Videos',
    labels={'views': 'Views', 'title': 'Video Title'},
    color_discrete_sequence=custom_colors
)

fig1.update_layout(
    xaxis_tickangle=-45,
    showlegend=True,
    plot_bgcolor='rgba(0,0,0,0)',
    paper_bgcolor='white'
)

fig1.show()
```

```python

import plotly.express as px

# Channel distribution
channel_counts = df['channelName'].value_counts().reset_index()
channel_counts.columns = ['channelName', 'count']

# Define custom color sequence
custom_colors = ['#FF6B6B', '#6BCB77', '#4D96FF', '#FFD93D', '#FF9CEE',
                 '#845EC2', '#00C9A7', '#FFC75F', '#F9F871', '#D65DB1',
                 '#FF9671', '#B0A8B9', '#0081CF', '#F76B8A', '#C34A36']

# Create pie chart with custom colors
fig2 = px.pie(
    channel_counts,
    names='channelName',
    values='count',
    title='📊 Number of Trending Videos per Channel',
    color_discrete_sequence=custom_colors
)

fig2.update_traces(textposition='inside', textinfo='percent+label')
fig2.update_layout(showlegend=True)

fig2.show()
```

```python

import pandas as pd
import plotly.express as px

trend_df = df.groupby(df['publishedDate'].dt.date)['views'].sum().reset_index()
trend_df.columns = ['Date', 'Total Views']

# Create line chart
fig3 = px.line(
    trend_df,
    x='Date',
    y='Total Views',
    title='📈 Total Views Over Time (Published Date)',
    markers=True
)

# Optional: Improve layout
fig3.update_layout(
    xaxis_title='Date',
    yaxis_title='Total Views',
    plot_bgcolor='white'
)

fig3.show()
```

```python

import pandas as pd
import plotly.express as px


# Step 1: Group by date and channel
multi_df = df.groupby([df['publishedDate'].dt.date, 'channelName'])['views'].sum().reset_index()
multi_df.columns = ['Date', 'Channel', 'Total Views']

# Step 2: Custom color palette
custom_colors = ['#FF6B6B', '#6BCB77', '#4D96FF', '#FFD93D', '#845EC2',
                 '#00C9A7', '#FFC75F', '#F9F871', '#D65DB1', '#FF9671']

# Step 3: Plot the line chart with hover
fig_multi = px.line(
    multi_df,
    x='Date',
    y='Total Views',
    color='Channel',
    title='📈 Views Over Time by Channel',
    color_discrete_sequence=custom_colors,
    markers=True,  # Add markers for clarity
    hover_data={
        'Date': True,
        'Channel': True,
        'Total Views': ':,'
    }
)

# Optional: Custom hover template (for precise control)
fig_multi.update_traces(
    hovertemplate='<b>Date:</b> %{x}<br><b>Channel:</b> %{legendgroup}<br><b>Views:</b> %{y:,}<extra></extra>'
)

# Step 4: Style the layout
fig_multi.update_layout(
    xaxis_title='Date',
    yaxis_title='Total Views',
    plot_bgcolor='white',
    legend_title='Channel Name'
)

fig_multi.show()
```

```python

import pandas as pd
import plotly.express as px

# Step 1: Get Top 5 Channels by Video Count
top_channels = df['channelName'].value_counts().head(5).index
filtered_df = df[df['channelName'].isin(top_channels)]

# Step 2: Group by Date and Channel
grouped_df = filtered_df.groupby([filtered_df['publishedDate'].dt.date, 'channelName'])['views'].sum().reset_index()
grouped_df.columns = ['Date', 'Channel', 'Total Views']

# Step 3: Define Custom Colors
custom_colors = ['#FF6B6B', '#6BCB77', '#4D96FF', '#FFD93D', '#845EC2']

# Step 4: Create Line Chart
fig = px.line(
    grouped_df,
    x='Date',
    y='Total Views',
    color='Channel',
    title='📈 Views Over Time – Top 5 Channels',
    color_discrete_sequence=custom_colors
)

# Step 5: Enhance Chart Appearance
fig.update_traces(mode="lines+markers")
fig.update_layout(
    xaxis_title='Date',
    yaxis_title='Total Views',
    plot_bgcolor='white',
    legend_title='Channel Name'
)

fig.show()
```

```python

channel_views = df.groupby('channelName')['views'].sum().reset_index().sort_values(by='views', ascending=False).head(10)

fig1 = px.bar(
    channel_views,
    x='channelName',
    y='views',
    title='🔥 Top 10 Channels by Total Views',
    color='channelName',
    text='views',
    color_discrete_sequence=px.colors.qualitative.Pastel
)
fig1.update_layout(xaxis_title='Channel', yaxis_title='Total Views')
fig1.show()
```

```python

df['DayOfWeek'] = df['publishedDate'].dt.day_name()
day_counts = df['DayOfWeek'].value_counts().reset_index()
day_counts.columns = ['Day', 'Video Count']

fig2 = px.bar(
    day_counts,
    x='Day',
    y='Video Count',
    title='📅 Trending Videos by Day of Week',
    color='Day',
    color_discrete_sequence=px.colors.qualitative.Safe
)
fig2.update_layout(xaxis_title='Day of Week', yaxis_title='Number of Videos')
fig2.show()
```

```python

import plotly.graph_objects as go

# Plotly line chart: Duration over Index
fig = go.Figure()

fig.add_trace(go.Scatter(
    y=df['duration'],
    mode='lines',
    line=dict(color='#00BFC4'),
    name='Duration'
))

# Update layout to match Matplotlib style
fig.update_layout(
    title='⏱️ Duration Over Index',
    xaxis_title='Index',
    yaxis_title='Duration',
    plot_bgcolor='white',
    showlegend=False,
    xaxis=dict(showline=True, linecolor='black', showgrid=True, gridcolor='lightgrey'),
    yaxis=dict(showline=True, linecolor='black', showgrid=True, gridcolor='lightgrey'),
)

fig.show()
```

```python

import plotly.graph_objects as go

# Plotly line chart: Views over Index
fig = go.Figure()

fig.add_trace(go.Scatter(
    y=df['views'],
    mode='lines',
    line=dict(color='#636EFA'),
    name='Views'
))

# Update layout to match Matplotlib style
fig.update_layout(
    title='📈 Views Over Index',
    xaxis_title='Index',
    yaxis_title='Views',
    plot_bgcolor='white',
    showlegend=False,
    xaxis=dict(showline=True, linecolor='black', showgrid=True, gridcolor='lightgrey'),
    yaxis=dict(showline=True, linecolor='black', showgrid=True, gridcolor='lightgrey'),
)

fig.show()
```

```python

import plotly.express as px

# Plotly scatter plot for Views vs Duration
fig = px.scatter(
    df,
    x='views',
    y='duration',
    title='📈 Views vs Duration',
    labels={'views': 'Views', 'duration': 'Duration'},
    opacity=0.8,                 # Similar to alpha=.8
    size_max=5,                  # Optional: control marker size
    color_discrete_sequence=['#EF553B']  # Optional color
)

# Customize layout to remove gridlines and borders (like Matplotlib)
fig.update_layout(
    plot_bgcolor='white',
    xaxis_title='Views',
    yaxis_title='Duration',
    showlegend=False
)

fig.update_xaxes(showline=True, linewidth=1, linecolor='black', gridcolor='lightgrey')
fig.update_yaxes(showline=True, linewidth=1, linecolor='black', gridcolor='lightgrey')

fig.show()
```

```python

# Only if 'duration' column is in seconds or numeric
fig3 = px.histogram(
    df,
    x='duration',
    nbins=50,
    title='⏱ Video Duration Distribution',
    color_discrete_sequence=['#FFA07A']
)
fig3.update_layout(xaxis_title='Duration (seconds)', yaxis_title='Video Count')
fig3.show()
```

```python

import plotly.figure_factory as ff

correlation = df[['views', 'duration']].corr()
fig4 = ff.create_annotated_heatmap(
    z=correlation.values,
    x=correlation.columns.tolist(),
    y=correlation.columns.tolist(),
    colorscale='Viridis'
)
fig4.update_layout(title_text='📈 Correlation Heatmap: Views vs Duration')
fig4.show()
```

```python

shorts_count = df['isShort'].value_counts().reset_index()
shorts_count.columns = ['Type', 'Count']
shorts_count['Type'] = shorts_count['Type'].map({True: 'Shorts', False: 'Regular'})

fig5 = px.pie(
    shorts_count,
    names='Type',
    values='Count',
    title='🎞 Shorts vs Regular Videos',
    color_discrete_sequence=px.colors.sequential.RdBu
)
fig5.show()
```

```python

upload_freq = df.groupby('channelName').agg({
    'videoId': 'count',
    'views': 'mean'
}).reset_index().rename(columns={'videoId': 'Video Count', 'views': 'Average Views'})

top_uploaders = upload_freq.sort_values(by='Video Count', ascending=False).head(15)

fig6 = px.scatter(
    top_uploaders,
    x='Video Count',
    y='Average Views',
    size='Average Views',
    color='channelName',
    hover_name='channelName',
    title='📊 Channel Upload Frequency vs Average Views',
    color_discrete_sequence=px.colors.qualitative.Set2
)
fig6.update_layout(xaxis_title='Video Count', yaxis_title='Average Views')
fig6.show()
```
